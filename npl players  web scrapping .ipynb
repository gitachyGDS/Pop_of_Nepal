```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
driver.get("https://cricketnepal.org.np/squads/nepal-mens-senior-national-team/")
```

```python
player_links=[]
for i in range(1,34):
    links=driver.find_element(By.XPATH,f"/html/body/section[3]/div/div/div[2]/div/div[1]/div/a[{i}]").get_attribute('href')

    player_links.append(links)
```

```python
driver=webdriver.Chrome()
players={}
for link in links[:3]:                          # Loop through first 3 links
    driver.get(link)                            # Open each link in browser
    time.sleep(3)      # Wait for page to fully load
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    
    driver.execute_script("window.scroll(0,0);")

    player_info=driver.find_element(By.XPATH,"/html/body/section[2]/div/div/div[2]/div/div[1]").text  # Locate player information section
    
    prof_info=driver.find_element(By.XPATH,"/html/body/section[2]/div/div/div[2]/div/div[2]").text   # Locate profile/professional information
        
    p_name=driver.title                         # Get page title (player name)
    players[p_name]={'player_info':player_info,'prof_info':prof_info}               
```
